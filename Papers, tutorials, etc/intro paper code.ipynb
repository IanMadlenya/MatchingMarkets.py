```python
# Arbitrary Match
mkt = mm.simulation(runs=5, time_per_run=3500, max_agents=50000, logAllData=False,
                   arrival_rate=3, 
                   crit_input=8, 
                   algorithm=mm.arbitraryMatch, 
                   metaAlgorithm=mm.meta_greedy, 
                   matchUtilFct=mm.utilSameType, 
                   success_prob=lambda: 1 if rng.rand() > 0.3 else 0)
mkt2 = deepcopy(mkt)
mkt2.metaAlgorithm=mm.meta_always
mkt3 = deepcopy(mkt)
mkt3.metaAlgorithm=mm.meta_patient
mkt4 = deepcopy(mkt)
mkt4.metaAlgorithm=mm.meta_periodic
mkt4.algoParams['period'] = 2
mkt5 = deepcopy(mkt4)
mkt5.algoParams['period'] = 3
mkt6 = deepcopy(mkt4)
mkt6.algoParams['period'] = 4


# Serial Dictatorship
mkt_1 = deepcopy(mkt)
mkt_1.algorithm = mm.serialDictatorship

mkt2_1 = deepcopy(mkt_1)
mkt2_1.metaAlgorithm=mm.meta_always
mkt3_1 = deepcopy(mkt2_1)
mkt3_1.metaAlgorithm=mm.meta_patient
mkt4_1 = deepcopy(mkt2_1)
mkt4_1.metaAlgorithm=mm.meta_periodic
mkt4_1.algoParams['period'] = 2
mkt5_1 = deepcopy(mkt4_1)
mkt5_1.algoParams['period'] = 3
mkt6_1 = deepcopy(mkt4_1)
mkt6_1.algoParams['period'] = 4


# RSD
mkt_2 = deepcopy(mkt)
mkt_2.algorithm = mm.serialDictatorship
mkt_2.algoParams['order'] = lambda x: rng.random()

mkt2_2 = deepcopy(mkt_2)
mkt2_2.metaAlgorithm=mm.meta_always
mkt3_2 = deepcopy(mkt2_2)
mkt3_2.metaAlgorithm=mm.meta_patient
mkt4_2 = deepcopy(mkt2_2)
mkt4_2.metaAlgorithm=mm.meta_periodic
mkt4_2.algoParams['period'] = 2
mkt5_2 = deepcopy(mkt4_2)
mkt5_2.algoParams['period'] = 3
mkt6_2 = deepcopy(mkt4_2)
mkt6_2.algoParams['period'] = 4
```

# Table 2

```python
sd = mm.simulation(runs=15, time_per_run=5500, max_agents=50000, logAllData=False,
                   arrival_rate=0.3, selfMatch=True,
                   crit_input=35, discount=lambda: 0.95,
                   algorithm=mm.serialDictatorship, 
                   metaAlgorithm=mm.meta_periodic, 
                   matchUtilFct=mm.utilRandom, 
                   success_prob=lambda: 1)

sd2 = deepcopy(sd)
sd2.algoParams['period'] = 3
sd3 = deepcopy(sd)
sd3.algoParams['period'] = 10
sd4 = deepcopy(sd)
sd4.algoParams['period'] = 35

# RSD
rsd = deepcopy(sd)
rsd.algoParams['order'] = lambda x: rng.random()

rsd2 = deepcopy(rsd)
rsd2.algoParams['period'] = 3
rsd3 = deepcopy(rsd)
rsd3.algoParams['period'] = 10
rsd4 = deepcopy(rsd)
rsd4.algoParams['period'] = 35

# TTC
ttc = deepcopy(sd)
ttc.algorithm = mm.TTC

ttc2 = deepcopy(ttc)
ttc2.algoParams['period'] = 3
ttc3 = deepcopy(ttc)
ttc3.algoParams['period'] = 10
ttc4 = deepcopy(ttc)
ttc4.algoParams['period'] = 35

# Max Weight Match
mwm = deepcopy(sd)
mwm.algorithm = mm.max_weight_matching

mwm2 = deepcopy(mwm)
mwm2.algoParams['period'] = 3
mwm3 = deepcopy(mwm)
mwm3.algoParams['period'] = 10
mwm4 = deepcopy(mwm)
mwm4.algoParams['period'] = 35

# Max Cardinality Matching
mcm = deepcopy(sd)
mcm.algorithm = mm.max_cardinality_matching

mcm2 = deepcopy(mcm)
mcm2.algoParams['period'] = 3
mcm3 = deepcopy(mcm)
mcm3.algoParams['period'] = 10
mcm4 = deepcopy(mcm)
mcm4.algoParams['period'] = 35
```

```python

sd.run()
sd2.run()
sd3.run()
sd4.run()

rsd.run()
rsd2.run()
rsd3.run()
rsd4.run()

ttc.run()
ttc2.run()
ttc3.run()
ttc4.run()

mwm.run()
mwm2.run()
mwm3.run()
mwm4.run()

mcm.run()
mcm2.run()
mcm3.run()
mcm4.run()
```
